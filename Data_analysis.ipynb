```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
history = np.genfromtxt('history.csv', delimiter=';')
history_val = np.genfromtxt('history_val.csv',delimiter=';')
z =  np.genfromtxt('z.csv',delimiter=';')
```

```python
history = history.reshape(6,6,20)
```

```python
history_new = np.zeros((6,20,6))

for i in range(6):
    history_new[i] = history[i].T
    
history = history_new
```

```python
Feed_temp = history[:,:,4] # Column Pressure
NT = history[:,:,2] + history[:,:,3] # Total number of trays
NR = history[:,:,2]  # Total above the feed
NS = history[:,:,3]  # Total below the feed
Water_Flash = history[:,:,5]
Z = history_val[:]
```

```python
x =  np.linspace(5,30,26)
y = np.linspace(400,450,51)
X,Y = np.meshgrid(x,y)


cp = plt.contour(x, y, z, levels =np.linspace(-2.67,-2.60,50))
plt.clabel(cp,fontsize=10)
plt.xlabel('Total number of stages')
plt.ylabel('Vaporised Feed kmol/h')
plt.show()
```

```python
plt.figure(figsize=(10,5))
for i in range(1,5): 
    plt.subplot(2,2,i)
    cp =plt.contour(x, y, z, levels =np.linspace(-2.68,-2.60,50))
    plt.clabel(cp,fontsize=8)
    plt.xlim([5, 25])
    plt.ylim([400,450])
    plt.plot(NS[i-1],Water_Flash[i-1], marker ='o', linestyle = 'none')
    plt.title("Iteration %d" %(i*5))

    
plt.tight_layout()
plt.show()
plt.savefig(fname="contour plot")
```
